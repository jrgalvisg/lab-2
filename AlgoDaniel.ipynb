```python
import matplotlib.pyplot as plt
%matplotlib inline
def swap(v, i, j):
    t = v[i]
    v[i] = v[j]
    v[j] = t
def perm(v, i, n):
    if i == n:
        v1 = v[:]
        permutaciones.append(v1)
    else:
        for j in range(i, n):
            swap(v, i, j)
            perm(v, i + 1, n)
            swap(v, i, j)
```

```python
def insertion_short(A):

    comparaciones = 0
    intercambios = 0
    instrucciones = 0
    for i in range(1,len(A)):
        temp = A[i]
        j = i - 1
        while j > -1 and temp < A[j]:
            comparaciones += 1
            A[j+1] = A[j]
            j = j-1
            intercambios += 1
            instrucciones += 3
        instrucciones += 1
        A[j+1] = temp
        instrucciones += 4

        if j != i:
            comparaciones += 1

    instrucciones += 1
    resultados = []
    resultados.append(instrucciones)
    resultados.append(comparaciones)
    resultados.append(intercambios)
    return resultados
```

```python
n = 2
A = list(i for i in range(n))
acum_ins = 0
acum_c = 0
acum_int = 0
long = 0
permutaciones = []
instrucciones = []
comparaciones = []
intercambios = []
perm(A, 0, n)

for a in range(len(permutaciones)):
    acum = insertion_short(permutaciones[a])
    acum_ins += acum[0]
    instrucciones.append(acum[0])
    acum_c += acum[1]
    comparaciones.append(acum[1])
    acum_int += acum[2]
    intercambios.append(acum[2])
    long += 1
print('ARREGLO DE TAMAÃ‘O {}'.format(n))
print('La cantidad de instrucciones promedio es: {}'.format(float(acum_ins / long)))
print('La cantidad de comparaciones promedio es: {}'.format(float(acum_c / long)))
print('La cantidad de intercambios promedio es: {}'.format(float(acum_int / long)))

bins = list(sorted(set(instrucciones)))
bins.append(bins[len(bins)-1]+2)
plt.hist(instrucciones, bins = bins,facecolor='blue', alpha=0.5, edgecolor = 'black',  linewidth=1)
plt.xlabel('NÃºmero de instrucciones')
plt.ylabel('Frecuencia')
plt.title('Instrucciones')
plt.show()

bins = list(sorted(set(comparaciones)))
bins.append(bins[len(bins)-1]+1)
plt.hist(comparaciones, bins = bins,facecolor='green', alpha=0.5, edgecolor = 'black',  linewidth=1)
plt.xlabel('NÃºmero de comparaciones')
plt.ylabel('Frecuencia')
plt.title('Comparaciones')
plt.show()

bins = list(sorted(set(intercambios)))
bins.append(bins[len(bins)-1]+1)
plt.hist(intercambios, bins = bins,facecolor='red', alpha=0.5, edgecolor = 'black',  linewidth=1)
plt.xlabel('NÃºmero de intercambios')
plt.ylabel('Frecuencia')
plt.title('Intercambios')
plt.show()
```
